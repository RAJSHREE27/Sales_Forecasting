```python
import numpy as np
import pandas as pd
from datetime import datetime

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Monthly_Sales_for_Laptops.csv', usecols=['Month', 'Sales', 'Budget'])

df['Sales'] = df['Sales'].replace('[\$,]', '', regex=True).astype(float)
df['Budget'] = df['Budget'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

```python
df.tail()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
df['Sales'] = df['Sales'].replace([np.inf, -np.inf], np.nan).fillna(0)
test_result=adfuller(df['Sales'].values)
```

```python
def adfuller_test(sales):
    result=adfuller(sales)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

```python
df
```

```python
#date format not changed and indexes is also not changed
adfuller_test(df['Sales'])
```

```python
#above p-value is higher than 0.05 as well as non-stationary
#trying sales first difference
df['Sales First Difference'] = df['Sales'] - df['Sales'].shift(1,fill_value=0)
adfuller_test(df['Sales First Difference'])
```

```python
#now the p-value is 0.27 and still the data is non-stationary
#trying sales second difference by shifting to only 1 position

df['Sales Second Difference'] = df['Sales First Difference'] - df['Sales First Difference'].shift(1,fill_value=0)
adfuller_test(df['Sales Second Difference'])
```

```python
#although p-value is higher but data is stationary. we can plot and check
df.plot(figsize=(15,8))
```

```python
import statsmodels.api as sm
```

```python
#Plots lags on the horizontal and the correlations on vertical axis. 
#lag value 30 and 60 used for better presentation
#we are going ahead with sales second difference since the data was stationary there
#ACF 

sm.graphics.tsa.plot_acf(df['Sales Second Difference'], lags=60)
plt.show()

sm.graphics.tsa.plot_acf(df['Sales Second Difference'], lags=30)
plt.show()
```

```python
#PACF

sm.graphics.tsa.plot_pacf(df['Sales Second Difference'], lags=60)
plt.show()
sm.graphics.tsa.plot_pacf(df['Sales Second Difference'], lags=30)
plt.show()
```

```python
#All the steps above where used to determine the order of model. So, difference(d)=2, p=12 and q=12 should be used. 
#But order(12,2,12) doesn't work and gives error.So lets try with other orders.

# p=1,d=2(as found earlier differnece of order 2 makes series stationary),q=1

model=sm.tsa.ARIMA(df['Sales'],order=(1,2,1))
model_fit=model.fit()
```

```python
print(model_fit.summary())
```

```python
# Plot residual errors

residuals = pd.DataFrame(model_fit.resid)
fig, ax = plt.subplots(1,2,figsize=(15, 8))
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()
```

```python
#The dynamic keyword affects in-sample prediction. 
#If dynamic is False, then the in-sample lagged values are used for prediction. 
#If dynamic is True, then in-sample forecasts are used in place of lagged dependent variables. 
#The first forecast value is start.

model_fit.plot_predict(dynamic=False)
plt.show()
```

```python
#splitting into train and test data
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iloc.html - reference

train=df['Sales'].iloc[:45]
test=df['Sales'].iloc[45:60]
```

```python
# model of order(1,2,1)
model=sm.tsa.ARIMA(train, order=(1, 2, 1))  
fitted=model.fit(disp=-1)  

fc, se, conf = fitted.forecast(15, alpha=0.05)

fc_series = pd.Series(fc, index=test.index)
lower_series = pd.Series(conf[:, 0], index=test.index)
upper_series = pd.Series(conf[:, 1], index=test.index)

plt.figure(figsize=(8,3), dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series, color='k', alpha=.15)
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```
